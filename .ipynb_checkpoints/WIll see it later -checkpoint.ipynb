```python
from gtts import gTTS
import os

welcome_text = "Namaste, welcome to your personal voice assistant. Please record your message."
tts = gTTS(welcome_text, lang="hi")   # Hindi accent
tts.save("welcome.mp3")
os.system("start welcome.mp3")  # macOS (use 'start' on Windows)
```

```python
import speech_recognition as sr

recognizer = sr.Recognizer()
with sr.Microphone() as source:
    print("Listening...")
    audio = recognizer.listen(source, timeout=5, phrase_time_limit=10)

try:
    user_prompt = recognizer.recognize_google(audio)
    print("User prompt:", user_prompt)
except sr.UnknownValueError:
    print("Could not understand audio")
```

```python
from transformers import GPT2Tokenizer, GPT2LMHeadModel

tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
model = GPT2LMHeadModel.from_pretrained("gpt2")
tokenizer.pad_token = tokenizer.eos_token
model.config.pad_token_id = tokenizer.pad_token_id

inputs = tokenizer(user_prompt, return_tensors="pt")
outputs = model.generate(
    inputs["input_ids"],
    attention_mask=inputs["attention_mask"],
    max_length=100,
    temperature=0.7,
    do_sample=True
)
response_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
print("Assistant response:", response_text)
```

```python

tts_response = gTTS(response_text, lang="en")
tts_response.save("response.mp3")
os.system("start response.mp3")  # macOS
```
