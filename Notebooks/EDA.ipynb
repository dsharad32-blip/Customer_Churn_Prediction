# Exploratory data Analysis
- Statistical Analysis
- Visualization


```python
#Imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
# Load the data 
CLEANED_DATAPATH=r"C:\Users\dshar\OneDrive\Pictures\Desktop\Customer_Churn_prediction\DATA\Clean_data.csv"
data=pd.read_csv(CLEANED_DATAPATH)
data.head()
```

```python
#Statistical Analysis
df=data.copy()
df.describe()
```

```python
#Categorical columns and Numerical columns 
cat_cols=df.select_dtypes(include="object").columns.to_list()
num_cols=df.select_dtypes(include="number").columns.to_list()
cat_cols.remove("Churn")
print("Categorical columns:\n",cat_cols)
print("\nNumerical columns:\n",num_cols)
```

```python
#Target Distribution
plt.figure(figsize=(10,6))
ax= sns.countplot(x="Churn",data=df)
#Bar label
for container in ax.containers:
    ax.bar_label(container)
```

```python
# Numeric col Vs. Churn
for col in num_cols:
    plt.figure(figsize=(10,6))
    sns.boxplot(x=col,hue="Churn",data=df)
    plt.ylabel(f"distribution of {col}")
    plt.xlabel("Count of Churn")
    plt.title(f"Churn Vs. {col}")
    plt.show()
```

```python

# Numeric col Vs. Churn
for col in num_cols:
    plt.figure(figsize=(10,6))
    sns.violinplot(x=col,hue="Churn",data=df)
    plt.ylabel(f"distribution of {col}")
    plt.xlabel("Count of Churn")
    plt.title(f"Churn Vs. {col}")
    plt.show()
```

```python
#Churn vs Categorical column 
for col in cat_cols:
    plt.figure(figsize=(10,6))
    ax=sns.countplot(x=col,hue="Churn",data=df)
    #Bar label 
    for container in ax.containers:
        ax.bar_label(container)
    plt.xlabel(f"{col}")
    plt.ylabel("Count")
    plt.title(f"{col} Vs.Churn")
    plt.xticks(rotation=40)
    plt.show()    
        
```

```python
#correlation 
corr=df.corr(numeric_only=True)
print("The correlation map :\n")
sns.heatmap(corr,annot=True)
plt.show()
```
