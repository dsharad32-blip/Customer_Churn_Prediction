# Data Pre-Processing 
- Data Type Coreection
- Duplicate Handling
- Remove irrelevant columns
- Save clean data

```python
#Imports
import pandas as pd 
import numpy as np
```

```python
#Loading data
DATA_PATH=r"C:\Users\dshar\OneDrive\Pictures\Desktop\Customer_Churn_prediction\DATA\Telco_Customer_Churn.csv"
data=pd.read_csv(DATA_PATH)
data.head()
```

```python
# copy 
df = data.copy()
df.info()
```

```python
#Meta data 
print("Number of rows:",df.shape[0])
print("Number of columns:",df.shape[1])
all_columns=df.columns.to_list()
num_columns=df.select_dtypes(include="number").columns.to_list()
cat_columns=df.select_dtypes(include="object").columns.to_list()
print("\nall_columns:\n",all_columns)
print("\nNumeric Columns:\n",num_columns)
print("\nCatergorical Columns",cat_columns)
print("\nDtypes:\n",df.dtypes)
```

```python
# Duplicate 
print("Duplicate details :\n",df.duplicated().sum())
print("Duplicate rows :\n",df.duplicated())
```

```python
# Drop Irrelvant column
df.drop(columns=['customerID'],axis=1,inplace=True)
```

```python
# Change the dtype of "'TotalCharges'"
df['TotalCharges']=pd.to_numeric(df['TotalCharges'],errors="coerce")
df['TotalCharges'].dtypes
```

```python
# Encode "Churn" column
df["Churn_encoded"]=df["Churn"].map({"Yes":1,"No":0})
df.head()
```

```python
#Save the clean data
Cleaned_data_path=r"C:\Users\dshar\OneDrive\Pictures\Desktop\Customer_Churn_prediction\DATA\Clean_data.csv"
df.to_csv(Cleaned_data_path,index=False)







```
